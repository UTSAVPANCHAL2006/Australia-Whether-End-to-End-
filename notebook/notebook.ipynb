```python
numerical_cols = []

categorical_cols = []


for col in df.columns:
    if df[col].dtype == 'object':
        categorical_cols.append(col)
    else:
        numerical_cols.append(col)
    
```

```python
for col in categorical_cols:
    mis = df[col].isnull().sum()
    pct = (mis / len(df)) * 100
    print(f"{col}, {mis}, {pct:.2f}%")
```

```python
for col in numerical_cols:
    mis = df[col].isnull().sum()
    pct = (mis / len(df)) * 100
    print(f"{col}, {mis}, {pct:.2f}%")
```

```python
for col in numerical_cols:
    df[col]= df[col].fillna(df[col].mean())
```

```python
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode().iloc[0]) 
```

```python
X = df.drop("RainTomorrow" , axis=1)
y = df["RainTomorrow"]

from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2)

from sklearn.preprocessing import LabelEncoder , StandardScaler

le_col = ['Location','WindGustDir','WindDir9am','WindDir3pm','RainToday']

for col in le_col:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col] = le.transform(X_test[col])
    

y_le = LabelEncoder()
y_train = y_le.fit_transform(y_train)
y_test = y_le.transform(y_test)


```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    scale_pos_weight=3.46,
    eval_metric='logloss',
    random_state=42)

xgb.fit(X_train,y_train)


y_test_pred = xgb.predict(X_test)
y_train_pred = xgb.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test = accuracy_score(y_test,y_test_pred)
train = accuracy_score(y_train,y_train_pred)
```

```python

print("Train accuracy:", train)
print("Test accuracy :", test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_test_pred))
```
